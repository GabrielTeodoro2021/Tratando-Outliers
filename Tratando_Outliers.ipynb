```python
import pandas as pd
import numpy as np
```

```python
casas = pd.read_csv('casas_para_alugar.csv')
```

```python
casas.head()
```

```python
casas.describe()
```

```python
casas['bathroom'].value_counts()
```

```python
# IQR = Distancia Inter Quartil (Inter Quantile Range)

q1 = casas['bathroom'].quantile(0.25) # Primeiro Quartil
q3 = casas['bathroom'].quantile(0.75) # Terceiro Quartil

iqr = q3 - q1

print(f'iqr: {iqr}')
```

```python
casas_outliers = casas[(casas['bathroom'] < q1 - (1.5 * iqr)) | (casas['bathroom'] > q3 + (1.5 * iqr))]
casas_outliers
```

```python
casas_outliers.describe()
```

```python
casas_inliers = casas[(casas['bathroom'] >= q1 - (1.5 * iqr)) & (casas['bathroom'] <= q3 + (1.5 * iqr))]
casas_inliers
```

```python
casas_inliers.describe()
```
